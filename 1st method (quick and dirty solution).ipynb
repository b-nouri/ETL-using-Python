```python
import boto3
import pandas as pd
from io import StringIO
```

# Read Data

```python
s3 = boto3.resource('s3')
bucket = s3.Bucket('xetra-1234')
bucket_obj1 = bucket.objects.filter(Prefix='dataset/2022-01-11/')
bucket_obj2 = bucket.objects.filter(Prefix='dataset/2022-01-12/')
```

```python
objects = [obj for obj in bucket_obj1] + [obj for obj in bucket_obj2]
objects
```

```python
df_all = pd.DataFrame()
for obj in objects:
    csv_obj = bucket.Object(key = obj.key).get().get('Body').read().decode('utf-8')
    data = StringIO(csv_obj)
    df = pd.read_csv(data,delimiter = ',')
    df_all = pd.concat([df_all,df])
df_all
```

```python
df_all.columns
```

```python
columns = ['ISIN','Date', 'Time', 'StartPrice', 'MaxPrice', 'MinPrice','EndPrice', 'TradedVolume', 'NumberOfTrades']
df_all = df_all[columns]
df_all.dropna(inplace=True)
df_all.columns = [x.lower() for x in df_all.columns]
```

```python
df_all
```

# Get Opening Price Per ISIN and Day

```python
df_all.reset_index(drop=True,inplace=True)
df_all['opening_price'] = df_all.sort_values('time').groupby(['isin','date'])['startprice'].transform('first')
df_all['closing_price'] = df_all.sort_values('time').groupby(['isin','date'])['endprice'].transform('last')
df_all = df_all.sort_values(['date','isin']).reset_index(drop=True)
df_all
```

# Aggregations

```python
df_all = df_all.groupby(['isin','date'],as_index=False).agg(opening_price_euro=('opening_price','min'),
                                                            closing_price= ('closing_price','min'),
                                                           minimun_price_euro= ('minprice','min'),
                                                           maximum_price_euro = ('maxprice','max'),
                                                           daily_traded_volume=('tradedvolume','sum'))
```

# Percent Change Prev Closing

```python
df_all['prev_closing_price'] = df_all.sort_values('date').groupby('isin')['closing_price'].shift(1)
df_all['percent_change_prev_closing'] = (df_all.closing_price - df_all.prev_closing_price) / df_all.prev_closing_price * 100
df_all.drop(columns = 'prev_closing_price',inplace=True)
```

```python
df_all = df_all.round(decimals = 2)
```
